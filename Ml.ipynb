```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data={
    "h_s":[5,6,7,4,3,2,8],
    "a_p":[89,65,78,54,89,50,97],
    "p_m":[55,72,87,45,32,30,96],
    "r":[1,1,1,1,0,0,1]
}
df=pd.DataFrame(data)
x=df[["h_s","a_p"]]
y=df["r"]
m=DecisionTreeClassifier()
m.fit(x,y)
new_s=[[2,70]]
print(m.predict(new_s))
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

data={
    "h_s":[5,6,7,4,3,2,8],
    "a_p":[89,65,78,54,89,50,97],
    "p_m":[55,72,87,45,32,30,96],
    "m":[60,67,85,43,36,68,98]
}
df=pd.DataFrame(data)
x=df[["h_s","a_p","p_m"]]
y=df["m"]
m=DecisionTreeRegressor()
m.fit(x,y)
new_s=[[8,70,45]]
print(m.predict(new_s))
```
